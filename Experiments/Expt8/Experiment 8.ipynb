```python
df11 = df2[df2.columns[1:]]
df21 = df2[df2.columns[:-1]]
```

### Country Data for k = 2

```python
model = KMeans(n_clusters=2, init = 'k-means++', max_iter=3)
model.fit(X1)
model.inertia_
```

```python
model.cluster_centers_
```

## Using the elbow method to find the optimal number of clusters

### For Country Data

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter=3, random_state = 42)
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### For credit data

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter=3, random_state = 42)
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
